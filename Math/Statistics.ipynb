## Maximum Likelihood Estimation

```python
x = np.random.normal(5,3,100)
```

```python
L = np.prod(sc_stats.norm.pdf(x, 5, 3), axis=0)
```

```python
for m in range(1,50):
    for s in range(1,50):
        l = np.prod(sc_stats.norm.pdf(x, m, s), axis=0)
        print(m,s)
        print(L > l)
```

## Simulating Gini purity likelihood of missclassification

```python
t = ['yes', 'no', 'yes', 'no', 'yes',  'no', 'yes']
rchoice = ['yes', 'no']
pchoice = [4/7,3/7]
```

```python
l_c = []
s_count = 0
for _ in range(1000):
    count = 0
    for e in t:
        if e != np.random.choice(rchoice, 1, p=pchoice):
            count +=1
            s_count +=1
    l_c.extend([count / 7 * 100])
#     print(count / 7 * 100)
print(np.mean(l_c), np.median(l_c), s_count / 7000 * 100)
```
