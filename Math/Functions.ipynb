```python
import numpy as np
import matplotlib.pyplot as plt

def main():
    N = 1000
    
    x = np.linspace(-100, 100, N)

    plt.figure()
    plt.plot(x, np.log(x))
    plt.xlabel('$x$')
    plt.ylabel('$\log(x)$')
    
    plt.figure()
    plt.plot(x, np.exp(x/10))
    plt.xlabel('$x$')
    plt.ylabel('$\exp(x)$')

    plt.figure()
    plt.plot(x, -np.exp(-x))
    plt.xlabel('$x$')
    plt.ylabel('$-\exp(-x)$')
    
    plt.figure()
    plt.plot(x, np.square(x))
    plt.xlabel('$x$')
    plt.ylabel('$x^2$')
    
    plt.figure()
    plt.plot(x, np.square(x))
    plt.xlabel('$x$')
    plt.ylabel('$x^2$')
    
    plt.figure()
    plt.plot(x, np.power(x,3))
    plt.xlabel('$x$')
    plt.ylabel('$x^3$')
    
    plt.show()

if __name__ == '__main__':
    main()
```
