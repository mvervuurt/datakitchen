# Model Building

```python
poverty_X = poverty_clean.drop(columns=['row_id','poverty_rate'], axis=1)
poverty_Y = poverty_clean.poverty_rate
```

## Feature Selection

```python
# use r2 adjusted in the future
scoring = {'r2':'r2','mse': make_scorer(mean_squared_error, greater_is_better=False)}
```

```python
cv = ShuffleSplit(n_splits=5, test_size=0.4, random_state=0)
```

```python
cachedir = mkdtemp()
estimators = [('scale_dim', RobustScaler()), ('reg', LinearRegression())]
regr_pipe = Pipeline(estimators, memory=cachedir)
```

```python
param_grid = dict(scale_dim=[None, StandardScaler(), MinMaxScaler(), MaxAbsScaler(), Normalizer(), RobustScaler()],
                  reg=[LinearRegression(), AdaBoostRegressor(DecisionTreeRegressor(max_depth=4),
                          n_estimators=300, random_state=0)])
```

```python
reg_grid = GridSearchCV(estimator=regr_pipe, param_grid=param_grid, scoring=scoring, cv=cv, error_score=0, refit='mse')
# reg_grid.fit(poverty_X, poverty_Y)
```

```python
train_scores = cross_validate(reg_grid, poverty_X, poverty_Y, cv=cv, scoring=scoring, return_train_score=True)
# np.average(np.sqrt(np.abs(train_scores['test_mse'])))
pprint(np.sqrt(np.abs(reg_grid.best_score_)))
pprint(reg_grid.best_estimator_)
```

```python
caching = mkdtemp()

mse = make_scorer(mean_squared_error, greater_is_better=False)

ada = AdaBoostRegressor(DecisionTreeRegressor(max_depth=4),
                          n_estimators=300, random_state=0)

# rfecv = RFECV(estimator=LinearRegression(normalize=False), step=1, cv=cv,
#               scoring=mse)

rfecv = RFECV(estimator=ada, step=1, cv=cv,
              scoring=mse)
rfecv.fit(poverty_X, poverty_Y)

res_rfecv = rfecv

# estimators = [('scale_dim', RobustScaler()), ('reg', rfecv)]
# regr_pipe = Pipeline(estimators, memory=caching)

# regr_pipe.fit(poverty_X, poverty_Y)

# res_rfecv = regr_pipe.named_steps['reg']

print("Optimal number of features : %d" % res_rfecv.n_features_)
```

```python
lasso_cv = LassoCV(cv=cv, normalize=True, random_state=0)

for k, (train, test) in enumerate(cv.split(poverty_X, poverty_Y)):
    lasso_cv.fit(poverty_X.iloc[train], poverty_Y.iloc[train])
    print("[fold {0}] alpha: {1:.5f}, score: {2:.5f}".
          format(k, lasso_cv.alpha_, lasso_cv.score(poverty_X.iloc[test], poverty_Y.iloc[test])))
```

```python
ridge_cv = RidgeCV(cv=cv, normalize=True, scoring='neg_mean_squared_error')

for k, (train, test) in enumerate(cv.split(poverty_X, poverty_Y)):
    ridge_cv.fit(poverty_X.iloc[train], poverty_Y.iloc[train])
    print("[fold {0}] alpha: {1:.5f}, score: {2:.5f}".
          format(k, ridge_cv.alpha_, ridge_cv.score(poverty_X.iloc[test], poverty_Y.iloc[test])))
```
